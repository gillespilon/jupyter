# KPIs

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
commits = pd.read_csv('kpis.csv')
commits
```

```python
#commits['total'] = commits.sum(axis=1)
#commits
commits['total']= commits.private + commits.support + commits.jupyter + commits.tableau
commits
```

```python
total_median = commits['total'].median()
total_xmin = commits['datenum'].min()
total_xmax = commits['datenum'].max()
```

```python
graphtitle = 'Key Performance Indicator'
graphsubtitle = 'Git Commits'
graphylabel = 'Number of commits'
graphxlabel = 'Day'
```

```python
# Create a new subplot from a grid of 1x1 (subplot)
# Enable moving spines (ax)
ax = plt.subplot(111)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.tick_params(direction='out')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# Plot private/ commits using the colour red with a continuous line of width 1 px
plt.plot(commits.datenum, commits.private, marker='o', color="red", linewidth=1.0, \
         linestyle="-", label="private/", )
# Plot support/ commits using the colour red with a continuous line of width 1 px
plt.plot(commits.datenum, commits.support, marker='o', color="blue", linewidth=1.0, \
         linestyle="-", label="support/")
for i, txt in enumerate( commits.supportnote ):
    ax.annotate( txt, ( commits.datenum[i] + .5, commits.support[i] + .5 ), color="blue" )
# Plot jupyter/ commits using the colour red with a continuous line of width 1 px
plt.plot(commits.datenum, commits.jupyter, marker='o', color="green", linewidth=1.0, \
         linestyle="-", label="jupyter/")
# Plot tableau/ commits using the colour red with a continuous line of width 1 px
plt.plot(commits.datenum, commits.tableau, marker='o', color="yellow", linewidth=1.0, \
         linestyle="-", label="tableau/")
# Set title and axis labels
plt.title(graphtitle + '\n' + graphsubtitle)
plt.ylabel(graphylabel)
plt.xlabel(graphxlabel)
ax.legend(loc='upper left', scatterpoints=1, frameon=False)
plt.savefig('kpi_commits.svg', format='svg')
plt.savefig('kpi_commits.pdf', format='pdf')
# Show the plot
plt.show()
```

```python
# Create a new subplot from a grid of 1x1 (subplot)
# Enable moving spines (ax)
ax = plt.subplot(111)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.tick_params(direction='out')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# Plot total commits using the colour blue with a continuous line of width 1 px
plt.plot(commits.datenum, commits.total, marker='o', color="blue", linewidth=1.0, \
         linestyle="-", label="total", )
# Plot median line for "total"
plt.plot([total_xmin, total_xmax], [total_median, total_median], color="red", \
         linewidth=1.0, \
         linestyle="-", label="median", )
# Set title and axis labels
plt.title(graphtitle + '\n' + graphsubtitle)
plt.ylabel(graphylabel)
plt.xlabel(graphxlabel)
plt.savefig('kpi_commits_total.svg', format='svg')
plt.savefig('kpi_commits_total.pdf', format='pdf')
# Show the plot
plt.show()
```
