# One-sample t test #

```python
import pandas as pd
import scipy.stats as stats
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
df = pd.read_csv('datafile_one_sample_t.csv', encoding='utf-16')
df
```

```python
y = df['y']
y
```

```python
%matplotlib inline
y.hist()
```

```python
# return_type='axes' to make it shut up about a future API change
y.to_frame().boxplot(return_type='axes')
```

```python
df.describe()
```

```python
stats.probplot(y, dist=stats.norm, plot=plt)
```

```python
target = 200
result = stats.ttest_1samp(y, target)
result
```

```python
average = np.mean(y)
if result.pvalue < 0.05:
    print('The average', average, 'of the sample is statistically, significantly ',  
          'different from the target', target, 'at the 0.05 significance level.')
else:
    print('The average', average, 'of the sample is not statistically, significantly ',  
          'different from the target', target, 'at the 0.05 significance level.')
```
