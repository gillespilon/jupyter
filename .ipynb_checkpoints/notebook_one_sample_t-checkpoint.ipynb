# One-sample t test #

## In brevi ##

A one-sample t test answers the question “is the average of a sample different from a specified value?”

**Example**. Your sample represents a current process. Is the average of this sample different from a known value, for example, a target (specification) or a previous time's average?

## How to use this notebook ##

1. Create a data file.
    - Name the column of your sample with the label **y**.
    - Save it as CSV (comma-separated values) with the name **datafile_one_sample_t.csv**.
    - Put it in the same directory as this notebook.
2. The value of the following variables can be changed to suit your data set.
    - **target**
    - **significance_level**
    - **confidence_level**

## Getting ready ##

Change the variables **target**, **significance_level**, and **confidence_level** to suit your analysis.

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display_html

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

target = 9.257
significance_level = 0.05
confidence_level = 0.95
```

## Read the data file ##

```python
df = pd.read_csv('datafile_one_sample_t.csv')
y = df['y'][df['y'].notnull()]
```

## Graphical analyses ##

### Histogram ###

```python
y.hist()
```

### Box plot ###

```python
# return_type='axes' to stop it from printing a note about a future API change
y.to_frame().boxplot(return_type='axes')
```

### Normal probability plot ###

```python
stats.probplot(y, dist=stats.norm, plot=plt)
```

## Confidence interval method ##

For a given confidence interval of an average of a sample, if many samples are collected and the confidence interval is calculated for each sample, in the long run about 95 % of these intervals would contain the true average.

The confidence limits are defined as:

$$
\begin{aligned}
    \overline{y}  \pm t_{1- \alpha / 2 \, \text{,} \, n-1} \times \frac{s}{\sqrt{n}} \\
\end{aligned}
$$

where:

$$
\begin{align*}
    \overline{y}                & = \text{the sample average} \\
    t_{1- \alpha / 2 \, \text{,} \, n-1} & = \text{the $100(1-\alpha/2)$ percentile of the t distribution with $n-1$ degrees of freedom} \\
    s                           & = \text{the sample standard deviation} \\
    n                           & = \text{the number of observations in the sample} \\
    \alpha                      & = \text{the desired significance level}
\end{align*}
$$

Reject H<sub>o</sub> if $\overline{y}$ is not within the confidence interval.

```python
n = len(y)
average = np.mean(y)
standard_deviation_sample = np.std(y,ddof=1)
t_critical = stats.t.isf(significance_level / 2, n - 1)
lower_limit = average - t_critical * standard_deviation_sample / np.sqrt(n)
upper_limit = average + t_critical * standard_deviation_sample / np.sqrt(n)
confidence_interval = lower_limit, upper_limit

if target < lower_limit or target > upper_limit:
    display_html('The sample average ' + str(average.round(3)) + ' of the sample is '
                 'statistically, significantly different from the target ' +
                 str(target) + ' because it is not within the interval ' +
                 '(' + str(lower_limit.round(3)) + ', ' + str(upper_limit.round(3)) + ').'
                 , raw=True)
else:
    display_html('The sample average ' + str(average.round(3)) + ' of the sample is not '
                 'statistically, significantly different from the target ' +
                 str(target) + ' because it is not within the interval ' +
                 '(' + str(lower_limit.round(3)) + ', ' + str(upper_limit.round(3)) + ').'
                 , raw=True)
```

## Hypothesis test method ##

To test whether the average of a sample is different from a known target, against the two-sided alternative that it is not different from a known target.

H<sub>o</sub>: average = target<br />
H<sub>a</sub>: average ≠ target<br />

The calculated value is defined as:

$$
\begin{aligned}
    t_\text{calculated} = \frac{\overline{y}-\text{target}}{s / {\sqrt{n}}} \\
\end{aligned}
$$

The critical value is defined as:

$$
\begin{aligned}
    t_\text{critical} = t_{1- \alpha / 2 \, \text{,} \, n-1} \\
\end{aligned}
$$

where:

$$
\begin{align*}
    \overline{y}                & = \text{the sample average} \\
    s                           & = \text{the sample standard deviation} \\
    n                           & = \text{the number of observations in the sample} \\
    t_{1- \alpha / 2 \, \text{,} \, n-1} & = \text{the $100(1-\alpha/2)$ percentile of the t distribution with $n-1$ degrees of freedom} \\
\end{align*}
$$

Reject H<sub>o</sub> if |t<sub>calculated</sub>| > t<sub>critical</sub>.

```python
t_calculated = (np.mean(y) - target) / (standard_deviation_sample / np.sqrt(n))
t_critical = stats.t.isf(significance_level / 2, n - 1)
p_value = stats.t.sf(np.abs(t_calculated), n - 1) * 2
display_html('H<sub>o</sub>: sample average = ' + str(target) + '<br />' +
             'H<sub>a</sub>: sample average ≠ ' + str(target) + '<br />' +
             't<sub>calculated</sub> = ' + str(t_calculated.round(3)) + '<br />' +
             't<sub>critical</sub> = ' + str(t_critical.round(3)) + '<br />' +
             'p value for t<sub>calculated</sub> = ' + str(p_value.round(3))
             , raw=True)

if abs(t_calculated) > t_critical:
    display_html('The average ' + str(average.round(3)) + ' of the sample is ' + 
             'statistically, significantly different from the target ' + str(target) + 
             ' because | t<sub>calculated</sub> | > t<sub>critical</sub> .', raw=True)
else:
    display_html('The average ' + str(average.round(3)) + ' of the sample is not ' + 
             'statistically, significantly different from the target ' + str(target) + 
             ' because | t<sub>calculated</sub> | < t<sub>critical</sub> .', raw=True)
```
