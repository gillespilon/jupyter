# Anscombe's quartet

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
anscombes_quartet_1 = pd.read_csv('anscombes_quartet_1.csv')
anscombes_quartet_2 = pd.read_csv('anscombes_quartet_2.csv')
anscombes_quartet_3 = pd.read_csv('anscombes_quartet_3.csv')
anscombes_quartet_4 = pd.read_csv('anscombes_quartet_4.csv')
```

```python
graphtitle = "Anscombe's Quartet"
graphylabel = 'Y'
graphxlabel = 'X'
```

```python
#Create the container, 1200 x 900 pixels
plt.figure(figsize=(12,9), dpi=100,frameon=True)
plt.suptitle(graphtitle, fontweight="bold")
ax1 = plt.subplot2grid((2,2), (0,0))
ax1.scatter(anscombes_quartet_1['x'], anscombes_quartet_1['y'], color="blue", \
           linewidth=0, linestyle="-", s=10, label="I")
#Calculate the linear regression model
m1, b1 = np.polyfit(anscombes_quartet_1['x'], anscombes_quartet_1['y'], 1)
ax1.plot(anscombes_quartet_1['x'], m1*anscombes_quartet_1['x'] + b1, '-', color="red", \
        label="Linear")
ax1.set_ylabel(graphylabel)
ax1.set_xlabel(graphxlabel)
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
ax1.tick_params(direction='out')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
#ax1.axis([2, 20, 2, 14])
ax1.set_ylim([2, 14])
ax1.set_xlim([2, 20])
#ax1.legend()
#plt.text(4, 13, 'I', fontsize=10)
ax1.set_title('Data set I', fontsize=10)
ax2 = plt.subplot2grid((2,2), (0,1))
ax2.scatter(anscombes_quartet_2['x'], anscombes_quartet_2['y'], color="blue", \
           linewidth=0, linestyle="-", s=10, label="II")
m2, b2 = np.polyfit(anscombes_quartet_2['x'], anscombes_quartet_2['y'], 1)
ax2.plot(anscombes_quartet_2['x'], m2*anscombes_quartet_2['x'] + b2, '-', color="red")
ax2.set_ylabel(graphylabel)
ax2.set_xlabel(graphxlabel)
ax2.spines['right'].set_color('none')
ax2.spines['top'].set_color('none')
ax2.tick_params(direction='out')
ax2.xaxis.set_ticks_position('bottom')
ax2.yaxis.set_ticks_position('left')
#ax2.axis([2, 20, 2, 14])
ax2.set_ylim([2, 14])
ax2.set_xlim([2, 20])
#ax2.legend()
#plt.text(4, 13, 'II', fontsize=10)
ax2.set_title('Data set II', fontsize=10)
ax3 = plt.subplot2grid((2,2), (1,0))
m3, b3 = np.polyfit(anscombes_quartet_3['x'], anscombes_quartet_3['y'], 1)
ax3.scatter(anscombes_quartet_3['x'], anscombes_quartet_3['y'], color="blue", \
           linewidth=0, linestyle="-", s=10, label="III")
ax3.plot(anscombes_quartet_3['x'], m3*anscombes_quartet_3['x'] + b3, '-', color="red")
ax3.set_ylabel(graphylabel)
ax3.set_xlabel(graphxlabel)
ax3.spines['right'].set_color('none')
ax3.spines['top'].set_color('none')
ax3.tick_params(direction='out')
ax3.xaxis.set_ticks_position('bottom')
ax3.yaxis.set_ticks_position('left')
#ax3.axis([2, 20, 2, 14])
ax3.set_ylim([2, 14])
ax3.set_xlim([2, 20])
#ax3.legend()
#plt.text(4, 13, 'III', fontsize=10)
ax3.set_title('Data set III', fontsize=10)
ax4 = plt.subplot2grid((2,2), (1,1))
ax4.scatter(anscombes_quartet_4['x'], anscombes_quartet_4['y'], color="blue", \
           linewidth=0, linestyle="-", s=10, label="IV")
m4, b4 = np.polyfit(anscombes_quartet_4['x'], anscombes_quartet_4['y'], 1)
ax4.set_ylabel(graphylabel)
ax4.set_xlabel(graphxlabel)
ax4.spines['right'].set_color('none')
ax4.spines['top'].set_color('none')
ax4.tick_params(direction='out')
ax4.xaxis.set_ticks_position('bottom')
ax4.yaxis.set_ticks_position('left')
#ax4.axis([2, 20, 2, 14])
ax4.set_ylim([2, 14])
ax4.set_xlim([2, 20])
#ax4.legend()
#plt.text(4, 13, 'IV', fontsize=10)
ax4.set_title('Data set IV', fontsize=10)
ax4.plot(anscombes_quartet_4['x'], m4*anscombes_quartet_4['x'] + b4, '-', color="red")
plt.savefig('anscombes_quartet.svg', format='svg')
plt.savefig('anscombes_quartet.pdf', format='pdf')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6), dpi=100)
plt.suptitle(graphtitle, fontweight="bold")
plt.ylabel(graphylabel)
plt.xlabel(graphxlabel)
ax = plt.subplot(111)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.tick_params(direction='out')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
m1, b1 = np.polyfit(anscombes_quartet_1['x'], anscombes_quartet_1['y'], 1)
plt.plot(anscombes_quartet_1['x'], m1*anscombes_quartet_1['x'] + b1, '-', color="red")
ax.scatter(anscombes_quartet_1['x'], anscombes_quartet_1['y'], color="blue", \
           linewidth=0, linestyle="-", s=10, label="I")
plt.axis([2, 20, 2, 14])
plt.text(4, 13, 'I', fontsize=10)
plt.ylabel(graphylabel)
plt.xlabel(graphxlabel)
plt.savefig('anscombes_quartet_1.svg', format='svg')
plt.savefig('anscombes_quartet_1.pdf', format='pdf')
plt.show()
```
