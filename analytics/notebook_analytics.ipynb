# Descriptive analytics

## Create a data file

- Name the columns of your data with the labels **month**, **product 1**, and **product 2**.
- Save it as CSV (comma-separated values)with the name **product_sales.csv**.
- Put it in the same directory as this notebook.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display_html

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

df = pd.read_csv('monthly_sales.csv')
df
```

## Descriptive statistics

```python
df.describe()
```

## Graphical analysis

```python
df[['product1']].hist()
df[['product2']].hist()
plt.show()
```

```python
# Create a new figure of size 8x6 points, using 100 dots per inch
# plt.figure(figsize=(8,6), dpi=100)

# Create a new subplot from a grid of 1x1
plt.subplot(111)

# Moving spines
ax = plt.subplot(111)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',75))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

# Plot y1 using the colour blue with a continuous line of width 1px
plt.plot(df.month, df.product1, color="blue", linewidth=1.0, linestyle="-", label="Product 1")

# Plot y2 using the colour red with a continuous line of width 1px
plt.plot(df.month, df.product2, color="red", linewidth=1.0, linestyle="-", label="Product 2")

# Set x limits
plt.xlim(0,19)

# Set x ticks
plt.xticks(np.linspace(1,18,18,endpoint=True))

# Set y limits
plt.ylim(75,525)

# Set y ticks
plt.yticks(np.linspace(100,500,5,endpoint=True))

# Set title and axis labels
plt.title('Monthly Sales for Products 1 and 2', fontweight='bold')
plt.ylabel('Sales (M USD)', fontweight='bold')
plt.xlabel('Month', fontweight='bold')

# Set a legend
plt.legend(loc='upper left', frameon=False)

# Save figure as svg
plt.savefig("monthly_sales_products_1_2.svg")

# Show the plot
plt.show()
```

# Predictive analytics

## Create a data file

- Name the columns of your data with the labels **machine_settings** and **energy_consumption**.
- Save it as CSV (comma-separated values)with the name **machine_settings.csv**.
- Put it in the same directory as this notebook.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

df = pd.read_csv('machine_settings.csv')
df
```

```python
# Create a new figure of size 8x6 points, using 100 dots per inch
# plt.figure(figsize=(8,6), dpi=100)

# Create a new subplot from a grid of 1x1
plt.subplot(111)

# Moving spines
ax = plt.subplot(111)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',-5))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',10))
plt.scatter(df.machine_setting,df.energy_consumption)

# Set x limits
plt.xlim(10,30)

# Set title and axis labels
plt.title('Energy Consumption v. Machine Setting', fontweight='bold')
plt.ylabel('Energy Consumption (M USD)', fontweight='bold')
plt.xlabel('Month', fontweight='bold')

# Save figure as svg
plt.savefig("machine_settings.svg")

# Show the plot
plt.show()
```

```python
z = np.polyfit(df.machine_setting,df.energy_consumption,3)
f = np.poly1d(z)
print(f)

#https://plot.ly/python/polynomial-fits/
```
